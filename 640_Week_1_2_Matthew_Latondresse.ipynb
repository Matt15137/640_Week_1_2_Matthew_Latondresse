```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/mattl/OneDrive/Documents/GitHub/DSC 640/ex1-2/hotdog-contest-winners.csv")
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Dogs eaten', data=df)
plt.title('Hot Dog Eating Contest Winners - Bar Chart')
plt.xlabel('Year')
plt.ylabel('Dogs eaten')
plt.show()
```

```python
# Stacked Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(df['Year'], df['Dogs eaten'], color='lightblue', label='Dogs eaten')
plt.bar(df['Year'], df['New record'], color='orange', label='New record')
plt.title('Hot Dog Eating Contest Winners - Stacked Bar Chart')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Dogs eaten'], labels=df['Year'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Hot Dog Eaten Per Year - Pie Chart')
plt.show()
```

```python
# Donut Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Dogs eaten'], labels=df['Year'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors, wedgeprops=dict(width=0.3))
plt.title('Hot Dog Eaten Per Year - Donut Chart')
plt.show()
```
